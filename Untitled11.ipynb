# **BUILDING A MODEL FOR WHEAT DISEASE DETECTION**

#**Importing the required libraries**


```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.vgg16 import preprocess_input
from tensorflow.keras.preprocessing import image
import tensorflow as tf
import numpy as np
```

# **Loading the dataset**
preprocessing the images and
labelling the images with the keras library

```python


train_datagen = ImageDataGenerator(
    rescale=1/255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2  # Assuming you have a validation set
)

train_generator = train_datagen.flow_from_directory(
    '/content/drive/MyDrive/train',  # Replace with the path to your dataset
    target_size=(200, 200),
    batch_size=32,
    class_mode='categorical',  # Use 'categorical' for one-hot encoded labels
    subset='training'
)

validation_generator = train_datagen.flow_from_directory(
    '/content/drive/MyDrive/validation',  # Replace with the path to your dataset
    target_size=(200, 200),
    batch_size=32,
    class_mode='categorical',  # Use 'categorical' for one-hot encoded labels
    subset='validation'
)

```

# **Checking the class indices**

```python
train_generator.class_indices
```

# **Defining the model**
Adding the layers to the model

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(16, (3, 3), activation='relu', input_shape=(200, 200, 3)),
    tf.keras.layers.MaxPool2D(2, 2),
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu'),
    tf.keras.layers.MaxPool2D(2, 2),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPool2D(2, 2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')
])

```

**Compile the model**

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

# **Model Training**
Training the model on three classes
Healthy ,Leaf rust and Stem rust images

```python
model.fit(train_generator, epochs=10, validation_data=validation_generator)
```

# **Predicting images with the trained model**
Preprocessing the images and predicting them with the trained model

```python
# Path to your image file
img_path = '/content/drive/MyDrive/test/leaf rust test.jpeg'

# Open and display the image using PIL
img = Image.open(img_path)
plt.imshow(img)
plt.axis('off')  # Optional: Turn off axis labels
plt.show()
# Load and preprocess the image

img = image.load_img(img_path, target_size=(200, 200))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
processed_img = preprocess_input(img_array)


# Make predictions
predictions = model.predict(processed_img)

# Interpret predictions (for classification)
predicted_class = np.argmax(predictions, axis=1)
print("Predicted class:", predicted_class)
if predicted_class == 1:
  print ("Disease Detected\n\tLeaf Rust")
elif predicted_class == 2:
  print ("Disease Detected\n\tStem Rust")
else:
  print("No Disease Detected\n\tHealthy Wheat")
```
